```python
from sklearn.model_selection import train_test_split
```

```python
X=X.reshape(-1,9)
X.shape
```

```python
df=pd.DataFrame(X,columns=attributes)
```

```python
df["Station"]=stations_column
```

```python
lookback = 5
master_arr=np.array([]).reshape(0,1,lookback,9)

for station in stations:
    station_arr=df.loc[df["Station"]==station].values
    station_arr=station_arr[:,:-1]

    new = np.lib.stride_tricks.sliding_window_view(x=station_arr, window_shape=(lookback, station_arr.shape[1]))
    master_arr=np.concatenate((master_arr,new))
X=master_arr
print(master_arr.shape)    
    
X = master_arr.reshape(master_arr.shape[0], lookback, 9)
print(X)
print(X.shape)
```

```python
y.shape
```

```python
y=y[lookback-1:]
```

```python
y.shape
```

```python
y=y.reshape(-1,)
```

```python
y.shape
```

```python
pd.Series(y).value_counts()
```

```python
class_weights={0:1,1:3.65}
```

```python
import keras
import tensorflow as tf
from numpy import unique
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers.core import Dense, Dropout
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
epochs = 100
batch_size = 32
n_hidden = 182

timesteps = lookback
input_dim = 9
n_classes = 2

model = Sequential()
model.add(LSTM(n_hidden, input_shape=(timesteps, input_dim),return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(int(n_hidden/2)))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=keras.optimizers.Adam(learning_rate=0.001),
              metrics=['accuracy'])
```

```python
X_train=X_train.astype(np.float32)
X_test=X_test.astype(np.float32)
```

```python
model.fit(X_train,
          y_train,
          batch_size=batch_size,
          validation_data=(X_test, y_test),
          epochs=epochs,
          class_weight=class_weights)
```
